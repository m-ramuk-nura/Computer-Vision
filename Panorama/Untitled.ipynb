```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import cv2
```

```python
im1 = cv2.imread(r'1.jpg')
im2 = cv2.imread(r'2.jpg')
im3 = cv2.imread(r'3.jpg')
```

```python
plt.subplot(131)
plt.imshow(im1[:,:,::-1])
plt.subplot(132)
plt.imshow(im2[:,:,::-1])
plt.subplot(133)
plt.imshow(im3[:,:,::-1])
plt.show()
```

```python
imgs = [im1,im2,im3]
```

```python
M = cv2.Stitcher.create(cv2.Stitcher_PANORAMA)
```

```python
status, panorama = M.stitch(imgs)
```

```python
plt.imshow(panorama[:,:,::-1])
```
