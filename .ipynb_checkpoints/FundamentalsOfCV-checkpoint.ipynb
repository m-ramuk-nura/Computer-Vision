```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'Resource/img.jpeg')
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
img = img[:,:,::-1]
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]
```

```python
plt.figure(1)
plt.subplot(231)
plt.imshow(img)
plt.xticks([])
plt.yticks([])

plt.subplot(232)
plt.imshow(img)
plt.xticks([])
plt.yticks([])

plt.subplot(233)
plt.imshow(img)
plt.xticks([])
plt.yticks([])

plt.subplot(234)
plt.imshow(R,cmap='grey')
plt.xticks([])
plt.yticks([])
plt.title('Red')

plt.subplot(235)
plt.imshow(G,cmap='grey')
plt.xticks([])
plt.yticks([])
plt.title('Green')

plt.subplot(236)
plt.imshow(B,cmap='grey')
plt.xticks([])
plt.yticks([])
plt.title('Blue')

plt.show()
```
